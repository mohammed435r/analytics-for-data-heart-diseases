```python

#outlier with boxplot
plt.figure(figsize=(20,10))
q1 = data['Cholesterol'].quantile(0.25)

q3 = data['Cholesterol'].quantile(0.75)

iqr = q3 - q1

lower = (q1 - 1.5 * iqr)

upper = (q3 + 1.5 * iqr)

print("Q1 (25%) =" ,q1 ,"   ","Q3 (75%) =" ,q3,"   ","IQR =" , iqr,"   ","lower =" ,lower ,"   ","upper =" ,upper)

outlier_array = []

number_of_outlier_element = 0

for item in data['Cholesterol']:

    if(item < lower):

        outlier_array.append(item)

        number_of_outlier_element = number_of_outlier_element + 1

    elif(item > upper):

        outlier_array.append(item)

        number_of_outlier_element = number_of_outlier_element + 1

outliersList = []

for i in outlier_array:

    if i not in outliersList:

        outliersList.append(i)

outliersList.sort()

print("Number Of Outlier Element : ",number_of_outlier_element,"elements")

print("Number Of Non Outlier Element : ",(data['Cholesterol'].count()-number_of_outlier_element),"elements")

print(f"Outliers is : {outliersList}")

sns.boxplot(data['Cholesterol'])
```

```python
from sklearn.neighbors import LocalOutlierFactor
plt.figure(figsize=(20,10))
x = data['MaxHR']

x=x.values.reshape(-1,2)

lof = LocalOutlierFactor(n_neighbors=13,contamination=0.03)

y_prediction = lof.fit_predict(x)

lofs_indx = np.where(y_prediction==-1)

Values = x[lofs_indx]

plt.scatter(x[:,0],x[:,1],color='k')

plt.scatter(Values[:,0],Values[:,1],color='r')

plt.show()

print(lofs_indx)
```
